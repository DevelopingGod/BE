#### Step 1: Import Libraries and Load Data
The IMDB dataset is built into Keras. We will load the top 10,000 most frequently used words. This keeps the data manageable while retaining the most important information.

```python
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten, GlobalAveragePooling1D

# Load the dataset
# num_words=10000 keeps the top 10,000 most frequently occurring words
print("Loading data...")
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")

# Check the first review (it's a list of integers, where each integer represents a word)
print("\nFirst review (as integers):", X_train[0])
```

#### Step 2: Data Preprocessing (Padding)
Neural networks expect inputs of the same length. Since movie reviews have different word counts, we must pad them to a fixed length (e.g., 250 words).

Too short? We add zeros.

Too long? We cut them off.

```python
max_review_length = 250

# Pad sequences ensures every review is exactly 250 integers long
X_train = sequence.pad_sequences(X_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length)

print(f"New shape of first review: {X_train[0].shape}")
```

#### Step 3: Build the Model
For text classification, we use an Embedding Layer first. This converts integer indices into dense vectors (meaningful representations of words).

1. Embedding Layer: Transforms our integer inputs into dense vectors of fixed size.

2. GlobalAveragePooling1D: Flattens the data by averaging the vector of each word. It's faster and less prone to overfitting than Flatten().

3. Dense (Hidden): Standard processing layer with ReLU.

4. Dense (Output): 1 Neuron with Sigmoid. This is critical for binary classification because it squashes the output between 0 and 1 (probability).

```python
embedding_vector_length = 32

model = Sequential()

# 1. Embedding Layer
# input_dim=10000 (vocabulary size), output_dim=32 (vector size), input_length=250
model.add(Embedding(input_dim=10000, output_dim=embedding_vector_length, input_length=max_review_length))

# 2. Flattening Strategy (GlobalAveragePooling is efficient for text)
model.add(GlobalAveragePooling1D())

# 3. Hidden Layer
model.add(Dense(16, activation='relu'))

# 4. Output Layer
# Sigmoid activation outputs a probability between 0 and 1
model.add(Dense(1, activation='sigmoid'))

# Compile the model
# binary_crossentropy is THE loss function for binary classification
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

#### Step 4: Train the Model
We train for a few epochs. Text models can overfit quickly, so we monitor the validation accuracy.

```python
history = model.fit(X_train, y_train,
                    validation_split=0.2,
                    epochs=5,
                    batch_size=32)
```

#### Step 5: Evaluate Accuracy
Check how the model performs on the test set.

```python
scores = model.evaluate(X_test, y_test, verbose=0)
print(f"Accuracy: {scores[1]*100:.2f}%")
```

#### Step 6: Test on Custom Reviews (Bonus)
The model actually works. We can decode the integers back to words or test a dummy prediction.

Since the mapping is complex to reverse-engineer in a simple block, here is how we interpret the output:

Output approaching 0: Negative Sentiment

Output approaching 1: Positive Sentiment

```python
# Simple manual check
prediction = model.predict(X_test[:5])
print("\nFirst 5 predictions:")
for i in range(5):
    sentiment = "Positive" if prediction[i] > 0.5 else "Negative"
    print(f"Review {i+1}: Prediction Score {prediction[i][0]:.4f} -> {sentiment} (Actual: {y_test[i]})")
```
